```python
from dotenv import load_dotenv
import os
from openai import OpenAI
load_dotenv()
# ===== CẤU HÌNH =====
API_KEY = os.getenv"OPENAI_API_KEY"  # 👈 Thay bằng API key của bạn
MODEL = "gpt-4o-mini"   # Hoặc "gpt-4o", "gpt-3.5-turbo"
SYSTEM_PROMPT = "Bạn là một trợ lý AI hữu ích, trả lời bằng tiếng Việt."

# ====================

client = OpenAI(api_key=API_KEY)
conversation_history = [{"role": "system", "content": SYSTEM_PROMPT}]


def chat(user_input):
    conversation_history.append({"role": "user", "content": user_input})

    response = client.chat.completions.create(
        model=MODEL,
        messages=conversation_history,
    )

    reply = response.choices[0].message.content
    conversation_history.append({"role": "assistant", "content": reply})
    return reply


def main():
    print("=" * 50)
    print("       🤖 ChatBot OpenAI - Terminal")
    print("   Gõ 'exit' hoặc 'quit' để thoát")
    print("   Gõ 'reset' để bắt đầu cuộc trò chuyện mới")
    print("=" * 50)

    while True:
        try:
            user_input = input("\n🧑 Bạn: ").strip()

            if not user_input:
                continue

            if user_input.lower() in ["exit", "quit"]:
                print("\n👋 Tạm biệt!")
                break

            if user_input.lower() == "reset":
                conversation_history.clear()
                conversation_history.append({"role": "system", "content": SYSTEM_PROMPT})
                print("✅ Đã reset cuộc trò chuyện.")
                continue

            print("\n🤖 Bot: ", end="", flush=True)
            reply = chat(user_input)
            print(reply)

        except KeyboardInterrupt:
            print("\n\n👋 Tạm biệt!")
            break
        except Exception as e:
            print(f"\n❌ Lỗi: {e}")


if __name__ == "__main__":
    main()
```
